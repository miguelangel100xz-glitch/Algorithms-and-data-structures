```python
def approx_vertex_cover(grafo):
  cubierto = set()
  aristas= _aristas(grafo)
  visitadas = set()

  for(u,v) in aristas:
    if u not in cubierto and v not in cubierto:
      cubierto.add(u)
      cubierto.add(v)
      visitadas.add((u,v))

  return cubierto


def _aristas(grafo):
  vistas = set()
  for u in grafo:
    for v in grafo[u]:
      arista = (min(u,v), max(u,v))
      if arista not in vistas:
        vistas.add(arista)
        yield arista


def verificar_cobertura(grafo, cobertura):
  for u in grafo:
    for v in grafo[u]:
      if u not in cobertura and v not in cobertura:
        return False
  return True

grafo = {
    1: {2, 3},
    2: {1, 4, 5},
    3: {1, 6},
    4: {2},
    5: {2, 6},
    6: {3, 5},
}

cobertura = approx_vertex_cover(grafo)
valida    = verificar_cobertura(grafo, cobertura)

print(f"Cobertura encontrada : {sorted(cobertura)}")
print(f"Tamaño               : {len(cobertura)}")
print(f"¿Válida?             : {valida}")
```
