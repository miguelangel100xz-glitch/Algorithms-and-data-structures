```python
import math
from typing import List, Tuple

Point = Tuple[float, float]

def euclidean(a: Point, b: Point) -> float:
  return math.hypot(a[0] - b[0], a[1] - b[1])

def dist_to_set(p: Point, centers: List[Point], dist=euclidean) -> float:
  return min(dist(p,c) for c in centers)

def center_selection(points: List[Point], k: int,
                     dist = euclidean) -> Tuple[List[Point], float]:
    if k >= len(points):
      return list(points), 0.0

    centers = [points[0]]
    for _ in range(k - 1):
      farthest = max(points, key=lambda p: dist_to_set(p, centers, dist))
      centers.append(farthest)

    radius = max(dist_to_set(p, centers, dist) for p in points)
    return centers, radius

def lower_bound_opt(points: List[Point], k: int,
                    dist = euclidean) -> float:

    centers = [points[0]]
    last_dist = 0.0
    for _ in range(k):
      d = max(dist_to_set(p, centers, dist) for p in points)
      farthest = max(points, key=lambda p: dist_to_set(p, centers, dist))
      centers.append(farthest)
      last_dist = d
    return last_dist / 2


if __name__ == "__main__":
    points: List[Point] = [
        (0, 0), (10, 0), (10, 10), (0, 10),
        (5, 5), (20, 0), (20, 10), (15, 5),
    ]
    k = 3

    centers, radius = center_selection(points, k)
    lb = lower_bound_opt(points, k)

    print(f"Puntos  : {points}")
    print(f"k       : {k}\n")
    print(f"Centros seleccionados:")
    for i, c in enumerate(centers):
        print(f"  c{i+1} = {c}")

    print(f"\nRadio obtenido : {radius:.4f}")
    print(f"Cota inf. OPT  : {lb:.4f}")
    print(f"Ratio real     : {radius / lb:.4f}  (garantía ≤ 2.0)")
```
