```python
def edit_distance(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m+1): dp[i][0] = i
    for j in range(n+1): dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(
                    dp[i-1][j],
                    dp[i][j-1],
                    dp[i-1][j-1]
                )

    return dp[m][n], dp



def backtrace(dp, s1, s2):
    i, j = len(s1), len(s2)
    ops = []
    while i > 0 or j > 0:
        if i > 0 and j > 0 and s1[i-1] == s2[j-1]:
            i -= 1; j -= 1
        elif i > 0 and j > 0 and dp[i][j] == dp[i-1][j-1]+1:
            ops.append(f"SUB  '{s1[i-1]}'→'{s2[j-1]}' en pos {i-1}")
            i -= 1; j -= 1
        elif i > 0 and dp[i][j] == dp[i-1][j]+1:
            ops.append(f"DEL  '{s1[i-1]}' en pos {i-1}")
            i -= 1
        else:
            ops.append(f"INS  '{s2[j-1]}' en pos {j-1}")
            j -= 1
    return list(reversed(ops))

dist, dp = edit_distance("kitten", "sitting")
print(f"Distancia: {dist}")          # 3
for op in backtrace(dp, "kitten", "sitting"):
    print(op)
```
