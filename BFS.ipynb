```python
from collections import deque

def bfs(grafo, inicio):
  """Recorre el grafo en anchura desde inicio"""
  visitados = set()
  cola = deque([inicio])
  visitados.add(inicio)
  orden = []


  while cola:
    nodo = cola.popleft()
    orden.append(nodo)

    for vecino in grafo[nodo]:
      if vecino not in visitados:
        visitados.add(vecino)
        cola.append(vecino)

  return orden

grafo = {
      1: [1,3],
      2: [1,4,5],
      3: [1,6,7],
      4:[2],
      5: [2],
      6: [3],
      7: [3],
  }


resultado = bfs(grafo, 1)
print(resultado)
```
