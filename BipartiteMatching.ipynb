```python
def bipartite_matching(graph):
  from collections import deque

  left_nodes = list(graph.keys())
  right_nodes = list({v for neighbors in graph.values() for v in neighbors})

  pair_left = {u: None for u in left_nodes}
  pair_right = {v: None for v in right_nodes}
  dist = {}

  def bfs():
    queue = deque()
    for u in left_nodes:
      if pair_left[u] is None:
        dist[u] = 0
        queue.append(u)
      else:
        dist[u] = float('inf')
    found = False

    while queue:
      u = queue.popleft()
      for v in graph.get(u, []):
        w = pair_right[v]
        if w is None:
          found = True
        elif dist.get(w, float('inf')) == float('inf'):
          dist[w] = dist[u] + 1
          queue.append(w)
    return found


  def dfs(u):
    for v in graph.get(u, []):
      w = pair_right[v]
      if w is None or (dist.get(w, float('inf')) == dist[u] + 1 and dfs(w)):
          pair_left[u] = v
          pair_right[v] = u
          return True
    dist[u] = float('inf')
    return False

  matching_size = 0
  while bfs():
    for u in left_nodes:
      if pair_left[u] is None:
        if dfs(u):
          matching_size += 1

  matching = {u: v for u,v in pair_left.items() if v is not None}
  return matching_size, matching

graph = {
    'W1': ['T1', 'T2'],
    'W2': ['T2', 'T3'],
    'W3': ['T3', 'T4'],
    'W4': ['T2', 'T4'],
}

size, matching = bipartite_matching(graph)
print(f"Matching maximo: {size}")
for worker, taks in matching.items():
  print(f"{worker} -> {taks}")
```
