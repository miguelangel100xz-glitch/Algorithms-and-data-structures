```python
import heapq

def interval_partitioning(intervals):
  if not intervals:
    return 0 , {}

  indexed = [(s,e,i) for i, (s , e) in enumerate(intervals)]
  sorted_intervals = sorted(indexed, key= lambda x: x[0])

  heap = []
  machine_count = 0
  assignment = {}

  for start, end, idx in sorted_intervals:
    if heap and heap[0][0] <= start:
      _, machine_id = heapq.heappop(heap)
    else:
        machine_count += 1
        machine_id = machine_count

    heapq.heappush(heap,(end, machine_id))
    assignment[idx] = machine_id

  return machine_count, assignment


intervals = [
    (0, 6),   # 0
    (1, 4),   # 1
    (3, 5),   # 2
    (5, 7),   # 3
    (3, 8),   # 4
    (6, 10),  # 5
    (8, 11),  # 6
    (9, 12),  # 7
]

num_machines, assignment = interval_partitioning(intervals)

print(f"Minimum machines/ rooms needed: {num_machines}\n")

from collections import defaultdict
machines = defaultdict(list)
for interval_idx, machine_id in assignment.items():
  machines[machine_id].append(intervals[interval_idx])

for machine_id in sorted(machines):
  slots = sorted(machines[machine_id])
  print(f"Machine {machine_id}: {slots}")

```
