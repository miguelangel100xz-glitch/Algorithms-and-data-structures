```python
import math

def distancia(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def fuerza_bruta(puntos):
    min_dist = float('inf')
    par = None
    n = len(puntos)
    for i in range(n):
        for j in range(i + 1, n):
            d = distancia(puntos[i], puntos[j])
            if d < min_dist:
                min_dist = d
                par = (puntos[i], puntos[j])
    return min_dist, par

def franja(puntos_franja, d):
    min_dist = d
    par = None
    puntos_franja.sort(key=lambda p: p[1])
    for i in range(len(puntos_franja)):
        j = i + 1
        while j < len(puntos_franja) and (puntos_franja[j][1] - puntos_franja[i][1]) < min_dist:
            d_actual = distancia(puntos_franja[i], puntos_franja[j])
            if d_actual < min_dist:
                min_dist = d_actual
                par = (puntos_franja[i], puntos_franja[j])
            j += 1
    return min_dist, par

def closest_pair_rec(puntos_x):
    n = len(puntos_x)
    if n <= 3:
        return fuerza_bruta(puntos_x)

    mid = n // 2
    punto_medio = puntos_x[mid]

    izq = puntos_x[:mid]
    der = puntos_x[mid:]

    d_izq, par_izq = closest_pair_rec(izq)
    d_der, par_der = closest_pair_rec(der)

    if d_izq < d_der:
        d_min, par_min = d_izq, par_izq
    else:
        d_min, par_min = d_der, par_der

    # Revisar franja central
    franja_puntos = [p for p in puntos_x if abs(p[0] - punto_medio[0]) < d_min]
    d_franja, par_franja = franja(franja_puntos, d_min)

    if d_franja < d_min:
        return d_franja, par_franja
    return d_min, par_min

def closest_pair(puntos):
    puntos_x = sorted(puntos, key=lambda p: p[0])
    return closest_pair_rec(puntos_x)


# Ejemplo
puntos = [(2, 3), (12, 30), (40, 50), (5, 1), (12, 10), (3, 4)]
distancia_min, par = closest_pair(puntos)
print(f"Par más cercano: {par[0]} y {par[1]}")
print(f"Distancia: {distancia_min:.4f}")
```
