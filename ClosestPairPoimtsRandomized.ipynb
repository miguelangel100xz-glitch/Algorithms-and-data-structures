```python
import random
import math

def dist(p1, p2):
    return math.hypot(p1[0]- p2[0], p1[1] - p2[1])


def closest_pair_randomizado(puntos):
    P = puntos[:]
    random.shuffle(P)

    mejor_dist = dist(P[0], P[1])
    mejor_par = (P[0], P[1])
    comparaciones = 1

    def cell(p, d):
      return (int(p[0] / d), int(p[1] / d))

    def reconstruir_grid(puntos_hasta_i, d):
      g = {}
      for p in puntos_hasta_i:
        c = cell(p, d)
        g.setdefault(c, []).append(p)
      return g

    grid = reconstruir_grid(P[:2], mejor_dist)

    for i in range(2, len(P)):
      p = P[i]
      gx, gy = cell(p, mejor_dist)

      for dx in range(-2,3):
        for dy in range(-2,3):
          vecinos = grid.get((gx + dx, gy +  dy), [])
          for q in vecinos:
            comparaciones += 1
            d = dist(p, q)

            if d < mejor_dist:
              mejor_dist = d
              mejor_par = (p,q)
              grid = reconstruir_grid(P[:1], mejor_dist)

      c = cell(p, mejor_dist)
      grid.setdefault(c, []).append(p)

    return mejor_par, mejor_dist, comparaciones


if __name__ =="__main__":
  import time

  n = 10_000
  puntos = [(random.uniform(0,1000), random.uniform(0,1000))
            for _ in range(n)]

  t0 = time.perf_counter()
  par, d, cmps = closest_pair_randomizado(puntos)
  t1 = time.perf_counter()

  bf = n * (n-1) // 2
  print(f"Par más cercano: {par[0]} ↔ {par[1]}")
  print(f"Distancia:       {d:.4f}")
  print(f"Comparaciones:   {cmps:,}  (fuerza bruta: {bf:,})")
  print(f"Reducción:       {(1 - cmps/bf)*100:.1f}%")
  print(f"Tiempo:          {(t1-t0)*1000:.1f} ms")
```
