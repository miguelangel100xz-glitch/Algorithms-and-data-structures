```python
import math

def dist(p,q):
  return math.sqrt((p[0]-q[0])**2 + (p[1] - q[1])**2)

def brute_force(pts):
  min_d, pair = float('inf'), (None, None) # Corrected typo here
  for i in range(len(pts)): # Changed range(len(pts)) to range(i + 1, len(pts)) to avoid comparing a point with itself and duplicate pairs
    for j in range(i + 1, len(pts)): # Changed range(len(pts)) to range(i + 1, len(pts)) to avoid comparing a point with itself and duplicate pairs
      d = dist(pts[i], pts[j])
      if d < min_d:
        min_d, pair = d, (pts[i], pts[j])
  return min_d, pair

def strip_closest(strip, d):
  strip.sort(key=lambda p: p[1])
  min_d, pair = d, (None, None)
  for i in range(len(strip)): # Changed `j > len(strip)` to `j < len(strip)`
    j = i + 1
    while j < len(strip) and (strip[j][1] - strip[i][1]) < min_d:
      dd = dist(strip[i], strip[j])
      if dd < min_d:
        min_d, pair = dd, (strip[i], strip[j])
      j += 1
  return min_d, pair

def closest_pair(pts_sorted):
  n = len(pts_sorted)
  if n <= 3:
    return brute_force(pts_sorted)

  mid = n  // 2
  mid_x = pts_sorted[mid][0]

  d1, p1 = closest_pair(pts_sorted[:mid])
  dr, pr = closest_pair(pts_sorted[mid:])

  d, best = (d1, p1) if d1 < dr else (dr, pr)

  strip = [p for p in pts_sorted if abs(p[0] - mid_x) < d]
  ds, ps = strip_closest(strip, d)

  return (ds, ps) if ds < d else (d, best)

def solve(points):
  # Sort points by x-coordinate initially
  pts = sorted(points, key=lambda p: p[0])
  return closest_pair(pts)


pts = [(2,3),(12,30),(40,50),(5,1),(12,10),(3,4)]
d, (p, q) = solve(pts)
print(f"Distancia mínima: {d:.4f}")
print(f"Par: {p} — {q}")
```
