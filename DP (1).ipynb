```python
def knapsack(pesos, valores, capacidad):
    n = len(pesos)
    dp = [[0] * (capacidad + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        peso_i = pesos[i - 1]
        valores_i = valores[i - 1]
        for w in range(capacidad + 1):
            if peso_i <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i-1][w - peso_i] + valores_i)
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacidad]


pesos = [2,3,4,5]
valores = [3,4,5,6]
capacidad = 5

print(knapsack(pesos, valores, capacidad))
```

```python
def knapsack_con_items(pesos, valores, capacidad):

    n = len(pesos)
    dp = [[0] * (capacidad + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(capacidad + 1):
            if pesos[i-1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i -1][w - pesos[i - 1]] + valores[i - 1])
            else:
              dp[i][w] = dp[i - 1][w]


    w = capacidad
    elegidos = []
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i - 1][w]:
            elegidos.append(i - 1)
            w -= pesos[i - 1]

    return dp[n][capacidad], elegidos[::-1]


```

```python
def knapstack_repeticion(pesos, valores, capacidad):
    n = len(pesos)
    dp = [0] * (capacidad + 1)

    for w in range(1, capacidad + 1):
        for i in range(n):
            if pesos[i] <= w:
                dp[w] = max(dp[w], dp[w - pesos[i]] + valores[i])

    return dp


pesos = [2,3,4,5]
valores = [3,4,5,6]
capacidad = 5

print(knapstack_repeticion(pesos, valores, capacidad))
```

```python
def knapsack_repeticion_con_items(pesos, valores, capacidad):
  n = len(pesos)
  dp = [0] * (capacidad + 1)
  eleccion = [-1] * (capacidad + 1)

  for w in range(1, capacidad + 1):
      for i in range(n):
          if pesos[i] <= w and dp[w - pesos[i]] + valores[i] > dp[w]:
              dp[w] = dp[w - pesos[i]] + valores[i]
              eleccion[w] = i

  w = capacidad
  conteo = {}
  while w > 0 and eleccion[w] != -1:
        i = eleccion[w]
        conteo[i] = conteo.get(i,0) + 1
        w -= pesos[i]

  return dp[capacidad], conteo



valor_max, objetos_usados = knapsack_repeticion_con_items(pesos, valores, capacidad)
print(f"Valor maximo: {valor_max}")
print(f"objetos usados {objetos_usados}")
```

```python
def knapsack_sin_repeticiones(pesos, valores, capacidad):
    n = len(pesos)
    dp = [0] * (capacidad  + 1)

    for i in range(n):
        for w in range(capacidad, pesos[i] - 1, - 1):
            dp[w] = max(dp[w], dp[w - pesos[i]] + valores[i])

    return dp[capacidad]



pesos = [2,3,4,5]
valores = [3,4,5,6]
capacidad = 5


print(knapsack_sin_repeticiones(pesos, valores, capacidad))
```

```python
def matrix_chain_order(dimensiones):
    num_dimensions = len(dimensiones) # For [10, 30, 5, 60], num_dimensions = 4
    num_matrices = num_dimensions - 1 # For 4 dimensions, there are 3 matrices

    # dp[i][j] stores the minimum multiplications for matrices i to j (1-based indexing)
    # The table size should accommodate indices up to num_matrices.
    dp = [[0] * (num_matrices + 1) for _ in range(num_matrices + 1)]

    # longitud is the length of the chain of matrices (from 2 up to num_matrices)
    for longitud in range(2, num_matrices + 1):
        # i is the starting index of the matrix chain (1-based)
        # i goes from 1 up to (num_matrices - longitud + 1)
        for i in range(1, num_matrices - longitud + 2):
            j = i + longitud - 1 # j is the ending index of the matrix chain

            dp[i][j] = float('inf')

            # k is the split point: (A_i...A_k) * (A_{k+1}...A_j)
            # k ranges from i to j-1
            for k in range(i, j):
                # Cost calculation: dp[i][k] + dp[k+1][j] + (p_{i-1} * p_k * p_j)
                # where p_x corresponds to dimensiones[x]
                costo = (dp[i][k] + dp[k + 1][j] +
                         dimensiones[i - 1] * dimensiones[k] * dimensiones[j])

                if costo < dp[i][j]:
                    dp[i][j] = costo

    # Return the minimum cost for the entire chain of matrices (from 1 to num_matrices)
    return dp[1][num_matrices], dp

dimensiones = [10, 30, 5, 60]

costo_minimo, dp = matrix_chain_order(dimensiones)
print(f"numero minimo de multiplicaciones: {costo_minimo}")
```

```python
def matrix_order_completo(dimensiones):
    n = len(dimensiones) - 1
    dp = [[0] * (n + 1) for _ in range(n + 1)]
    division = [[0] * (n + 1) for _ in range(n + 1)]

    for longitud in range(2, n + 1):
        for i in range(1, n - longitud + 2):
            j = i + longitud - 1
            dp[i][j] = float('inf')
            for k in range(i, j):
                costo = (dp[i][k] + dp[k + 1][j] +
                         dimensiones[i - 1] * dimensiones[k] * dimensiones[j]) # Changed ** to *
                if costo < dp[i][j]:
                    dp[i][j] = costo
                    division[i][j] = k

    return dp[1][n], division


def construir_parentizacion(division, i, j):
    if i == j:
        return f"A{i}"

    k = division[i][j]
    izquierda = construir_parentizacion(division, i, k)
    derecha = construir_parentizacion(division, k + 1, j) # Fixed typo 'paretizacion'

    return f"({izquierda} x {derecha})"


dimensiones = [10,30,5,60]
costo_minimo, division = matrix_order_completo(dimensiones) # Fixed function name
orden_optimo = construir_parentizacion(division, 1, len(dimensiones) - 1) # Fixed function name

print(f"costo minimo: {costo_minimo}")
print(f"orden optimo: {orden_optimo}")
```

```python
def floyd_warshall(grafo):
    n = len(grafo)
    dp = [fila[:] for fila in grafo]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dp[i][k] + dp[k][j] < dp[i][j]:
                    dp[i][j] = dp[i][k] + dp[k][j]

    return dp


INF = float('inf')

grafo = [
    [0,   3,   INF, 7],
    [8,   0,   2,   INF],
    [5,   INF, 0,   1],
    [2,   INF, INF, 0]
]



distancias = floyd_warshall(grafo)

for fila in distancias:
    print(fila)
```

```python
import heapq

def dijkstra(graph, origen):
    # Use the 'graph' parameter, not a global 'grafo'
    dist = {nodo: float('inf') for nodo in graph}
    dist[origen] = 0
    visitados = set()
    cola = [(0, origen)]

    while cola:
        d_actual, nodo_actual = heapq.heappop(cola)

        if nodo_actual in visitados:
            continue
        visitados.add(nodo_actual)

        # Use 'graph' parameter and fix typo 'nodoo_actual' to 'nodo_actual'
        for vecino, peso in graph[nodo_actual]:
            nueva_dist = d_actual + peso
            if nueva_dist < dist[vecino]:
                dist[vecino] = nueva_dist
                heapq.heappush(cola, (nueva_dist, vecino))

    # Correct indentation for return statement
    return dist


grafo = {
    'A': [('B', 3), ('D', 7)],
    'B': [('C', 2)],
    'C': [('D', 1)],
    'D': []
}

print(dijkstra(grafo, 'A'))
```
