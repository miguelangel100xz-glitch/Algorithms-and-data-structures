```python
def mergesort(arr):
    if  len(arr) <= 1:
        return arr

    medio = len(arr) // 2
    izquierda = mergesort(arr[:medio])
    derecha = mergesort(arr[medio:])

    return merge(izquierda, derecha)


def merge(izquierda, derecha):
    resultado = []
    i = j = 0

    while i < len(izquierda) and j < len(derecha):
      if izquierda[i] <= derecha[j]:
          resultado.append(izquierda[i])
          i += 1
      else:
          resultado.append(derecha[j])
          j  += 1

    resultado.extend(izquierda[i:])
    resultado.extend(derecha[j:])

    return resultado


numero = [38, 27, 43, 3, 9, 82, 10]
print(mergesort(numero))
```

```python
def multiplicar_matrices(A,B):
    filas_A = len(A)
    columnas_A = len(A[0])
    filas_B = len(B)
    columnas_B = len(B[0])

    if columnas_A != filas_B:
        raise ValueError("error")

    resultado = [[0 for _ in range(columnas_B)] for _ in range(filas_B)]

    for i in range(filas_A):
        for j in range(columnas_B):
            for k in range(columnas_A):
                resultado[i][j] += A[i][k] * B[k][j]

    return resultado


A = [[1,5,7],
     [7,2,9]]


B = [[4,7],
     [9,3],
     [52,19]]


print(multiplicar_matrices(A,B))
```

```python
import cmath

def fft(x):
    n = len(x)

    if n <= 1:
        return x

    if n % 2 != 0:
        raise ValueError("Error")

    pares = fft(x[0::2])
    impares = fft(x[1::2])


    resultado = [0] * n
    for k in range(n // 2):
        factor = cmath.exp(-2j * cmath.pi * k / n)
        resultado[k] = pares[k] + factor * impares[k]
        resultado[k + n // 2] = pares[k] - factor * impares[k]

    return resultado


señal = [1,1,1,1,0,0,0,0]
transformada = fft(señal)

for i, val in enumerate(transformada):
    print(f"X[{i}] = {val:.4f}")
```
