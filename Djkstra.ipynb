```python
import heapq

def dijkstra(grafo, inicio):

  distancias = {n: float('inf') for n in grafo}
  predecesores = {n: None for n in grafo}
  distancias[inicio] = 0

  heap = [(0, inicio)]

  while heap:
    dist_u,u = heapq.heappop(heap)

    if dist_u > distancias[u]:
      continue

    for(peso, v) in grafo.get(u, []):
      nueva_dist = distancias[u] + peso

      if nueva_dist < distancias[v]:
        distancias[v] = nueva_dist
        predecesores[v] = u
        heapq.heappush(heap, (nueva_dist,v))

  return distancias, predecesores


def reconstruir_ruta(predecesores, inicio, destino):

  ruta, nodo = [], destino
  while nodo is not None:
    ruta.append(nodo)
    nodo = predecesores[nodo]
  ruta.reverse()
  return ruta if ruta[0] == inicio else []

grafo = {
    'A': [(2, 'B'), (6, 'C')],
    'B': [(2, 'A'), (1, 'C'), (5, 'D')],
    'C': [(6, 'A'), (1, 'B'), (3, 'D'), (4, 'E')],
    'D': [(5, 'B'), (3, 'C'), (1, 'E')],
    'E': [(4, 'C'), (1, 'D')],
}

dists, preds = dijkstra(grafo, 'A')

print("Distancias mínimas desde A:")
for nodo, d in dists.items():
    ruta = reconstruir_ruta(preds, 'A', nodo)
    print(f"  A → {nodo}: costo {d}  |  ruta: {' → '.join(ruta)}")
```
