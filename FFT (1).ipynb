```python
import numpy as np
def fft_cooley_tukey(x):
    N = len(x)

    if N <= 1:
        return x

    pares = fft_cooley_tukey(x[0::2])
    impares = fft_cooley_tukey(x[1::2])

    T = [np.exp(-2j * np.pi * k /N) * impares[k] for k in range(N//2)]

    return [pares[k] + T[k]   for k in range(N // 2)] + \
           [pares[k] - T[k]   for k in range(N // 2)]



if __name__ == '__main__':
    N = 128
    t = np.linspace(0,1,N, endpoint= False)
    señal = np.sin(2* np.pi * 5 * t) + 0.5 * np.sin(2* np.pi * 20 * t)

    X = fft_cooley_tukey(list(señal))
    magnitud = [abs(v) for v in X]
    X_numpy = np.fft.fft(señal)
    error = np.max(np.abs(np.array(X) - X_numpy))

    print(f"Error maximo vs numpy: {error:.2e}")

    freqs = np.fft.fftfreq(N, d=1/N)
    for i, m in enumerate(magnitud[:N//2]):
        if m > 5:
          print(f"frecuencia detectada: {freqs[i]:.0f} Hz (magnitud = {m:.1f})")
```

```python
from ast import Tuple
from collections import defaultdict

class Grafo:
  def __init__(self):
    self.adyacencia = defaultdict(list)

  def agregar_arista(self,u,v):
    self.adyacencia[u].append(v)
    self.adyacencia[v].append(u)


def dfs_recursivo(grafo, nodo, visitados = None):
    if visitados is None:
        visitados = set()

    visitados.add(nodo)
    print(f"Visitando: {nodo}")

    for vecino in grafo.adyacencia[nodo]:
        if vecino not in visitados:
            dfs_recursivo(grafo, vecino, visitados)

    return visitados



def dfs_iterativo(grafo, inicio):
    visitados = set()
    pila = [inicio]


    while pila:
      nodo = pila.pop()

      if nodo not in visitados:
        visitados.add(nodo)
        print(f"visitando: {nodo}")

        for vecino in grafo.adyacencia[nodo]:
            if vecino not in visitados:
                pila.append(vecino)

    return visitados



def dfs_ciclo(grafo, nodo, visitados=None, en_pila=None):
    if visitados is None:
        visitados = set()
        en_pila = set()

    visitados.add(nodo)
    en_pila.add(nodo)

    for vecino in grafo.adyacencia[nodo]:
        if vecino not in visitados:
            if dfs_ciclo(grafo, vecino, visitados, en_pila):
                return True
        elif vecino in en_pila:
            return True
    en_pila.remove(nodo)
    return False



def dfs_topo(grafo, nodo, visitados, orden_topo):
    visitados.add(nodo)

    for vecino in grafo.adyacencia[nodo]:
      if vecino not in visitados:
          dfs_topo(grafo, vecino, visitados, orden_topo)

    orden_topo.append(nodo)


def orden_topologico(grafo,nodos):
    visitados = set()
    pila_topo = []

    for nodo in nodos:
        if nodo not in visitados:
            dfs_topo(grafo, nodo, visitados, pila_topo)

    return pila_topo[::-1]


if __name__ == '__main__':
    g = Grafo()
    for u,v in [("A","B"), ("A","C"), ("B","D"),("C","D"), ("D","E")]:
        g.agregar_arista(u,v)

    print("== DFS RECURSIVO")
    dfs_recursivo(g, "A")

    print("== DFS ITERATIVO")
    dfs_iterativo(g, "A")

    print("== DFS CICLO")
    dfs_ciclo(g, "A")

    print("== DFS ORDEN TOPOLOGICO")
    print(orden_topologico(g, ["A","B","C","D","E"]))
```
