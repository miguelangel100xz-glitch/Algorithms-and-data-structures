```python
def floyd_warshall(grafo):
    n = len(grafo)
    dist = [fila[:] for fila in grafo]
    next_vertex = [[None] * n for _ in range(n)]


    for i in range(n):
        for j in range(n):
            if i != j and dist[i][j] != float('inf'):
              next_vertex[i][j] = j

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    next_vertex[i][j] = next_vertex[i][k]

    for i in range(n):
        if dist[i][i] < 0:
            raise ValueError("El grafo contiene un ciclo de peso negativo")

    return dist, next_vertex


def reconstruir_camino(next_vertex, inicio, fin):
    if next_vertex[inicio][fin] is None:
        return None

    camino = [inicio]
    while inicio != fin:
        inicio = next_vertex[inicio][fin]
        camino.append(inicio)
    return camino


if __name__ == "__main__":

    INF = float('inf')

    grafo = [
        [0,   5,   INF, 10],
        [INF, 0,   3,   INF],
        [INF, INF, 0,   1],
        [INF, INF, INF, 0]
    ]


    dist, next_vertex = floyd_warshall(grafo)

    print("Matriz de distancias minimas: ")
    for fila in dist:
        print([x if x != INF else 'inf' for x in fila])


    print("\nCamino mas corto 0 a 3: ")
    print(reconstruir_camino(next_vertex, 0, 3))

    print("Distancia 0 a 3: " , dist[0][3])
```
