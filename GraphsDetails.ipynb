```python
def dfs(graph, start):
    visited = set()
    order = []

    def _dfs(node):
        visited.add(node)
        order.append(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                _dfs(neighbor)

    _dfs(start)
    return order


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}


print(dfs(graph, 'A'))


```

```python
def dfs_dirigido(graph, start):
    visited = set()
    order = []
    ciclos = []

    def _dfs(node, stack_set):
        visited.add(node)
        stack_set.add(node)
        order.append(node)

        for neighbor in graph[node]:
            if neighbor in stack_set:
                ciclos.append((node, neighbor))
            elif neighbor not in visited:
                _dfs(neighbor, stack_set)

        stack_set.remove(node)

    _dfs(start, set())
    return order, ciclos

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

orden , ciclos = dfs_dirigido(graph, 'A')
print("orden de visita:", orden)
print("ciclos detectados:", ciclos)

```

```python
def dfs_dag_topo(graph):
    visited = set()
    finished = []

    def _dfs(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                _dfs(neighbor)
        finished.append(node)

    for node in graph:
        if node not in visited:
            _dfs(node)

    return list(reversed(finished))


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['E', 'G'],
    'D': ['F'],
    'E': ['F', 'G'],
    'F': ['H'],
    'G': ['H'],
    'H': [],
}


print(dfs_dag_topo(graph))

```

```python
from collections import defaultdict

def kosaraju(graph):
    visited = set()
    finish_order = []

    def dfs1(node):
        visited.add(node)
        for nb in graph[node]:
            if nb not in visited:
                dfs1(nb)
        finish_order.append(node)

    for node in graph:
        if node not in visited:
            dfs1(node)


    transposed = defaultdict(list)
    for u in graph:
        for v in graph[u]:
            transposed[v].append(u)

    visited2 = set()
    sccs = []


    def dfs2(node, comp):
        visited2.add(node)
        comp.append(node)
        for nb in transposed[node]:
            if nb not in visited2:
                dfs2(nb, comp)

    for node in reversed(finish_order):
        if node not in visited2:
            comp = []

            dfs2(node, comp)
            sccs.append(comp)

    return sccs


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['E', 'G'],
    'D': ['F'],
    'E': ['F', 'G'],
    'F': ['H'],
    'G': ['H'],
    'H': [],
}


for i, scc in enumerate(kosaraju(graph),1):
    print(f"SCC {i}: {scc}")
```

```python
from IPython.core.interactiveshell import dis
from collections import defaultdict, deque

def bfs(graph, start):
    visited = {start}
    queue = deque([start])
    dist = {start: 0}
    order = []

    while queue:
        node = queue.popleft()
        order.append(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                dist[neighbor] = dist[node] + 1
                queue.append(neighbor)

    return order, dist


graph = {
    'A' : ['B', 'C'],
    'B' : ['D', 'E'],
    'C' : ['F'],
    'D' : [],
    'E' : ['F'],
    'F' : []
}


orden, distancias = bfs(graph, 'A')
print("odern:", orden)
print("Distancias:", distancias)
```

```python
from collections import deque

def camino_mas_corto(graph, start, end):
    visited = {start}
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == end:
            return path

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path + [neighbor])

    return None

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}


print(camino_mas_corto(graph, 'A', 'F'))
```

```python
import heapq

def djkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    dist[start] = 0
    prev = {}
    heap = [(0, start)]

    while heap:
        d, node = heapq.heappop(heap)

        if d > dist[node]:
            continue

        for neighbor, weight in graph[node]:
            new_dist = dist[node] + weight
            if new_dist < dist[neighbor]:
                dist[neighbor] = new_dist
                prev[neighbor] = node
                heapq.heappush(heap, (new_dist, neighbor))

    return dist, prev


def reconstruir_camino(prev, start, end):
    path = []
    node = end
    while node != start:
        path.append(node)
        node = prev[node]

    path.append(start)
    return list(reversed(path))


graph = {
    'A': [('B',4), ('C',2)],
    'B': [('A',4), ('D',5), ('E',1)],
    'C': [('A',2), ('D',8), ('F',10)],
    'D': [('B',5), ('C',8), ('E',2), ('F',6)],
    'E': [('B',1), ('D',2), ('G',3)],
    'F': [('C',10), ('D',6), ('G',5)],
    'G': [('E',3), ('F',5)],
}


dist, prev = djkstra(graph, 'A')
print("Distancias:", dist)
print("camino a -> g", reconstruir_camino(prev, 'A', 'G'))
```
