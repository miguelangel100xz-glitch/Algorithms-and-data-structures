```python
from heapq import heappush, heappop


EDGES = [
    (4,  'A', 'B'),
    (8,  'A', 'C'),
    (11, 'A', 'D'),
    (2,  'A', 'F'),
    (6,  'B', 'C'),
    (10, 'B', 'E'),
    (5,  'C', 'D'),
    (7,  'C', 'E'),
    (9,  'D', 'E'),
    (1,  'D', 'G'),
    (3,  'F', 'G'),
]

NODES = ['A', 'B', 'C', 'D', 'E', 'F', 'G']


class UnionFind:

    def __init__(self, nodes):
        self.parent = {n: n for n in nodes}
        self.rank = {n: 0 for n in nodes}

    def find(self,x):
        if self.parent[x] != x:
          self.parent[x] = self.find(self.parent[x])
        return self.parent[x]


    def union(self, x, y):
        rx, ry = self.find(x), self.find(y)
        if rx == ry:
            return False

        if self.rank[rx] < self.rank[ry]:
            rx, ry = ry, rx
        self.parent[ry] = rx
        if self.rank[rx] == self.rank[ry]:
            self.rank[rx] += 1
        return True


def kruskal(nodes, edges):
    uf = UnionFind(nodes)
    mast = []
    total = 0

    for peso, u, v in sorted(edges):
            if uf.union(u,v):
              mast.append((peso, u, v)) # Changed mst to mast
              total += peso
              print(f"{u}-{v}  pesos = {peso} total acum.={total}")
            else:
                print(f"{u}-{v} peso = {peso}   total aum.={total}")

            if len(mast) == len(nodes) - 1: # Changed mst to mast
                break
    return mast, total # Changed mst to mast




def build_adjacency(nodes, edges):
    adj = {n: [] for n in nodes}
    for peso, u, v in edges:
        adj[u].append((peso, v))
        adj[v].append((peso, u))
    return adj


def prim(nodes, edges, start='A'):
    adj = build_adjacency(nodes, edges)
    visitados = {start}
    heap = []
    mst = []
    total = 0

    for peso, vecino in adj[start]:
        heappush(heap, (peso, start, vecino))

    while heap and len(mst) < len(nodes) - 1:
        peso, u, v = heappop(heap)

        if v in visitados:
            print(f"{u}-{v}  peso = {peso} {v} ya visitado")
            continue

        visitados.add(v)
        mst.append((peso, u, v))
        total += peso
        print(f"{u}-{v} peso = {peso} total aum.={total}")


        for p_vecino, vecino in adj[v]:
            if vecino not in visitados:
                heappush(heap, (p_vecino, v, vecino))

    return mst, total




def imprimir_mst(nombre, mst, total):
    print(f"\n{'═'*40}")
    print(f"  MST — {nombre}")
    print(f"{'═'*40}")
    for peso, u, v in mst:
        print(f"  {u} ── {v}   peso {peso}")
    print(f"  {'─'*24}")
    print(f"  Costo total: {total}")
    print(f"{'═'*40}\n")


if __name__ == "__main__":

    print("\n" + "─"*40)
    print("  KRUSKAL — pasos")
    print("─"*40)
    mst_k, total_k = kruskal(NODES, EDGES)
    imprimir_mst("Kruskal", mst_k, total_k)

    print("─"*40)
    print("  PRIM — pasos (inicio: A)")
    print("─"*40)
    mst_p, total_p = prim(NODES, EDGES, start='A')
    imprimir_mst("Prim", mst_p, total_p)
```

```python
import heapq
from collections import Counter


class Nodo:
    def __init__(self, char, freq):
      self.char = char
      self.freq = freq
      self.left = None
      self.right = None

    def __lt__(self, otro):
      return self.freq < otro.freq

    def __repr__(self):
        return f"Nodo({self.char}, {self.freq})"


def contruir_arbol(texto: str) -> Nodo:
    frecuencias = Counter(texto)
    heap = [Nodo(char, freq) for char, freq in frecuencias.items()]
    heapq.heapify(heap)

    print("Frecuencias iniciales:")
    for n in sorted(heap, key=lambda x: x.freq):
        print(f"  {n.char}: {n.freq}")
    print()


    paso = 1
    while len(heap) > 1:
        izq = heapq.heappop(heap)
        der = heapq.heappop(heap)

        padre = Nodo(char=None, freq=izq.freq + der.freq)
        padre.left = izq
        padre.right = der

        heapq.heappush(heap, padre)
        print(f"Paso {paso}: unir {izq} + {der} Nodo(freq={padre.freq})")
        paso += 1
    return heap[0]


def generar_codigos(nodo:Nodo, prefijo: str = "", tabla: dict = None) -> dict:
    if tabla is None:
        tabla = {}

    if nodo is None:
        return tabla


    if nodo.char is not None:
        tabla[nodo.char] = prefijo or "0" # Corrected the syntax error here
        return tabla

    generar_codigos(nodo.left, prefijo + "0", tabla)
    generar_codigos(nodo.right, prefijo + "1", tabla)
    return tabla


def codificar(texto: str, tabla: dict) -> str:
    return "".join(tabla[c] for c in texto)



def decodificar(bits: str, raiz: Nodo) -> str:
    resultado = []
    nodo = raiz
    for bit in bits:
        nodo = nodo.left if bit == "0" else nodo.right

        if nodo.char is not None:
            resultado.append(nodo.char)
            nodo = raiz

    return "".join(resultado)



def estadisticas(texto:str, bits: str, tabla: dict):
    bits_originales = len(texto) * 8
    bits_comprimidos = len(bits)
    ahorro = (1- bits_comprimidos / bits_originales) * 100

    print(f"\n{'═'*42}")
    print(f"  Tabla de códigos Huffman")
    print(f"{'═'*42}")
    print(f"  {'Char':<6} {'Freq':>5}   {'Código':<12} {'Bits':>4}")
    print(f"  {'─'*38}")
    freq = Counter(texto)
    for char, code in sorted(tabla.items(), key=lambda x: len(x[1])):
        print(f"  '{char}'    {freq[char]:>5}   {code:<12} {len(code):>4}")
    print(f"{'═'*42}")
    print(f"\n  Texto original : {texto!r}")
    print(f"  Bits originales: {bits_originales}  ({len(texto)} chars × 8 bits)")
    print(f"  Bits Huffman   : {bits_comprimidos}")
    print(f"  Ahorro         : {ahorro:.1f}%")
    print(f"\n  Bits codificados:")
    print(f"  {bits}")



if __name__ == "__main__":
    texto = "ABRACADABRA"
    print(f"{'─'*42}")
    print(f"  Construyendo árbol para: {texto!r}")
    print(f"{'─'*42}\n")

    raiz  = contruir_arbol(texto)
    tabla = generar_codigos(raiz)
    bits  = codificar(texto, tabla)
    texto_recuperado = decodificar(bits, raiz)

    estadisticas(texto, bits, tabla)

    print(f"\n  Decodificado   : {texto_recuperado!r}")
    assert texto == texto_recuperado, "¡Error: el texto decodificado no coincide!"
    print(f"  Verificación   : OK — texto original recuperado correctamente\n")

```
