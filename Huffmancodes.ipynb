```python
import heapq
from collections import Counter

class Nodo:
  def __init__(self, simbolo, frecuencia):
    self.simbolo = simbolo
    self.frecuencia = frecuencia
    self.izquierdo = None
    self.derecho = None

  def __lt__(self, otro): # Moved __lt__ to be a method of the class
    return self.frecuencia < otro.frecuencia


def construir_arbol(texto):
  frecuencias = Counter(texto)

  heap = [Nodo(s,f) for s, f in frecuencias.items()]
  heapq.heapify(heap)

  while len(heap) > 1:
    izq = heapq.heappop(heap)
    der = heapq.heappop(heap)

    padre = Nodo(None, izq.frecuencia + der.frecuencia)
    padre.izquierdo = izq
    padre.derecho = der

    heapq.heappush(heap, padre)

  return heap[0]


def generar_codigos(nodo, prefijo="" , codigos=None):
  if codigos is None:
    codigos = {}

  if nodo is None:
    return codigos

  if nodo.simbolo is not None:
    codigos[nodo.simbolo] = prefijo if prefijo else "0"
    return codigos

  generar_codigos(nodo.izquierdo, prefijo + "0", codigos)
  generar_codigos(nodo.derecho, prefijo + "1", codigos)
  return codigos

def codificar(texto, codigos):
  return "".join(codigos[c] for c in texto)

def decodificar(bits, raiz):
  resultado = []
  nodo = raiz
  for bit in bits:
    nodo = nodo.izquierdo if bit == "0" else nodo.derecho
    if nodo.simbolo is not None:
      resultado.append(nodo.simbolo)
      nodo = raiz
  return "".join(resultado)


if __name__ == "__main__":
    texto = "hola mundo"

    raiz   = construir_arbol(texto)
    codigos = generar_codigos(raiz)

    bits_original = len(texto) * 8
    codificado    = codificar(texto, codigos)
    decodificado  = decodificar(codificado, raiz)

    print("Texto original :", texto)
    print("Decodificado   :", decodificado)
    print(f"Bits originales: {bits_original}")
    print(f"Bits Huffman   : {len(codificado)}")
    print(f"Ahorro         : {100*(1 - len(codificado)/bits_original):.1f}%")
    print("\nTabla de códigos:")
    for simbolo, codigo in sorted(codigos.items(), key=lambda x: len(x[1])):
        s = repr(simbolo)
        print(f"  {s:6} \u2192 {codigo:10}  (freq {Counter(texto)[simbolo]})")
```
