```python
def interval_scheduling(intervals):
  sorted_intervals = sorted(intervals, key= lambda x: x[1])
  selected = []
  last_end = float('-inf')
  for start, end in sorted_intervals:
    if start >= last_end:
      selected.append((start,end))
      last_end = end
  return selected

intervals = [
    (1, 4),
    (3, 5),
    (0, 6),
    (5, 7),
    (3, 8),
    (5, 9),
    (6, 14),
    (8, 9),
    (8, 12),
    (2, 13),
    (12, 14),
]


result = interval_scheduling(intervals)
print("Selected intervals", result)
print("count: ", len(result))
```

```python
def weighted_interval_scheduling(intervals):
  import bisect

  jobs = sorted(intervals, key=lambda x: x[1])
  n = len(jobs)
  ends = [j[1] for j in jobs]

  def latest_compatible(i):
    idx = bisect.bisect_right(ends, jobs[i][0], 0 , i) - 1
    return idx

  dp = [0] * (n+1)
  for i in range(1, n + 1):
      job = jobs[i - 1]
      p = latest_compatible(i - 1)
      dp[i] = max(dp[i - 1], dp[p + 1] + job[2])

  selected = []
  i = n
  while i > 0:
    job = jobs[i - 1]
    p = latest_compatible(i - 1)
    if dp[p + 1] + job[2] > dp[i - 1]:
      selected.append(job)
      i = p + 1
    else:
      i -= 1
  return dp[n], list(reversed(selected))

weighted = [(1, 4, 3), (3, 5, 4), (0, 6, 2), (5, 9, 5), (6, 10, 3), (8, 11, 5)]
max_w, chosen = weighted_interval_scheduling(weighted)
print("Max weight:", max_w)
print("Chosen:", chosen)

```
