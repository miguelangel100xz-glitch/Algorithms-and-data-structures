```python
def karatsuba(x,y):
  if x < 10 or y < 10:
    return x*y

  n = max(len(str(x)), len(str(y)))
  m = n // 2

  factor = 10 ** m
  x_hi, x_lo = divmod(x, factor)
  y_hi, y_lo = divmod(y, factor)

  z0 = karatsuba(x_lo, y_lo)
  z2 = karatsuba(x_hi, y_hi)
  z1 = karatsuba(x_lo + x_hi, y_lo + y_hi)

  middle = z1 - z2 - z0

  return z2 * (10 ** (2*m)) + middle * (10**m) + z0


print(karatsuba(1234, 5678))
print(karatsuba(644664,874532))
```
