```python
def knapstack(pesos: list[int], valores: list[int], W: int) -> tuple[int, list[int]]:

  n = len(pesos)

  dp = [[0] * (W + 1) for _ in range(n + 1)]

  for i in range(1, n + 1):
    for w in range(W + 1):
      dp[i][w] = dp[i - 1][w]
      if w >= pesos[i - 1]:
        incluir = dp[i - 1][w - pesos[i - 1]] + valores[i - 1]
        if incluir > dp[i][w]:
          dp[i][w] = incluir

  seleccionadas = []

  w = W
  for i in range(n, 0, -1):
    if dp[i][w] != dp[i - 1][w]:
      seleccionadas.append(i - 1)
      w -= pesos[i - 1]

  return dp[n][W], list(reversed(seleccionadas))


duraciones = [3,5,2,7,4]
prioridades = [4,5,3,7,6]
W = 9

valor_max, idx = knapstack(duraciones, prioridades, W)
print(f"Valor maximo: {valor_max}")
print(f"Tareas seleccionadas: {[f'T{i+1}(dur={duraciones[i]}, val={prioridades[i]})' for i in idx]}")
print(f"Tiempo usado: {sum(duraciones[i] for i in idx)}h / {W}h")
```
