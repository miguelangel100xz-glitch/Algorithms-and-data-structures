```python
def encontrar(padre,x):
  if padre[x] != x:
    padre[x] = encontrar(padre,padre[x])
  return padre[x]

def unir(padre,rango,x,y):
  rx,ry = encontrar(padre,x), encontrar(padre,y)
  if rx == ry:
    return False

  if rango[rx] < rango[ry]:
    rx, ry = ry,rx
  padre[ry] = rx
  if rango[rx] == rango[ry]:
    rango[rx] += 1
  return True



def kruskal(nodos,aristas):
  """
  """
  aristas_ord = sorted(aristas)

  padre = {n: n for n in nodos}
  rango = {n: 0 for n in nodos}

  aristas_mst = []
  peso_total = 0

  for(w,u,v) in aristas_ord:
    if unir(padre, rango, u, v):
      aristas_mst.append((w,u,v))
      peso_total += w
      if len(aristas_mst) == len(nodos) -1:
        break

  return aristas_mst, peso_total


nodos = ['A','B','C','D','E']
aristas = [
    (1, 'B', 'C'),
    (2, 'A', 'B'),
    (3, 'A', 'C'),
    (4, 'B', 'D'),
    (5, 'C', 'D'),
]

mst, peso_total = kruskal(nodos, aristas)

print("Arbol de expansion minima:")
for(w,u,v) in mst:
  print(f"{u} - {v} : {w}")

print(f"Peso total: {peso_total}")
```
