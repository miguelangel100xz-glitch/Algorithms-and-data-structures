```python
def least_squares_error(points, i, j):

  pts = points[i:j +1]
  n = len(pts)
  if n == 1:
    return 0.0
  sx = sum(p[0] for p in pts)
  sy = sum(p[1] for p in pts)
  sxy = sum(p[0]*p[1] for p in pts)
  sxx = sum(p[0]**2 for p in pts)
  denom = n*sxx - sx*sx
  if abs(denom) < 1e-12:
    return sum((p[1] - sy/n)**2 for p in pts)
  a = (n*sxy - sx*sy) / denom
  b = (sy - a*sx) / n
  return sum((p[1] - (a*p[0] + b))**2 for p in pts)

def segmented_least_squares(points, C):
  n = len(points)
  pts = sorted(points)

  e = [[0.0]*n for _ in range(n)]
  for i in range(n):
    for j in range(1,n):
      e[i][j] = least_squares_error(pts, i, j)

  OPT = [float('inf')] * n
  seg = [0] * n

  for j in range(n):
    for i in range(j+1):
      prev = OPT[i-1] if i > 0 else 0
      cost = e[i][j] + C + prev
      if cost < OPT[j]:
        OPT[j] = cost
        seg[j] = i

  segments, J = [] , n - 1
  while j >= 0:
    i = seg[j]
    segments.append((i,j))
    j = i - 1
  segments.reverse()
  return segments, OPT[n-1]

# Ejemplo
pts = [(1,1),(2,3),(3,2),(5,8),(6,10),(8,5),(9,4),(10,6)]
segs, cost = segmented_least_squares(pts, C=5)
for i, j in segs:
    print(f"Segmento [{i}..{j}]: puntos {pts[i:j+1]}")
print(f"Costo total: {cost:.2f}")
```
