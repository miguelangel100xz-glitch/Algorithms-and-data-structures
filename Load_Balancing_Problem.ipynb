```python
import heapq
from typing import List, Tuple

def load_balancing_lpt(jobs: List[float], m: int) -> Tuple[float, List[List[float]]]:

  sorted_jobs = sorted(enumerate(jobs), key=lambda x: -x[1])
  heap = [(0.0,i) for i in range(m)]
  heapq.heapify(heap)

  assignment = [[] for _ in range(m)]
  for job_id, pj in sorted_jobs:
    load, machine = heapq.heappop(heap)
    assignment[machine].append((job_id, pj))
    heapq.heappush(heap, (load + pj, machine))

  makespan = max(load for load, _ in heap)
  return makespan, assignment

def lower_bound(jobs: List[float], m: int) -> float:
  return max(max(jobs), sum(jobs) / m)


if __name__ == "__main__":
  jobs = [8,7,6,5,4,3,2,1]
  m = 3

  makespan, assignment = load_balancing_lpt(jobs, m)
  lb = lower_bound(jobs, m)

  print(f"Maquinas : {m}")
  print(f"Treas : {jobs}\n")

  for i, tasks in enumerate(assignment):
    ids = [f"J{jid}" for jid, _ in tasks]
    total = sum(p for _, p in tasks)
    print(f"Maquina {i+1}: {ids}  -> carga = {total}")


  print(f"\nMakespan obtenido : {makespan}")
  print(f"Cota inferior OPT : {lb:.2f}")
  print(f"Ratio aproximación: {makespan / lb:.4f}")
```
