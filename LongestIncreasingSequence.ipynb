```python
from bisect import bisect_left

def lis_with_sequence(nums):
    tails = []
    indices = []
    predecessors = [-1] * len(nums)
    tails_idx = []

    for i, x in enumerate(nums):
        pos = bisect_left(tails, x)
        if pos == len(tails):
            tails.append(x)
            tails_idx.append(i)
        else:
            tails[pos] = x
            tails_idx[pos] = i
        predecessors[i] = tails_idx[pos - 1] if pos > 0 else -1

    seq = []
    idx = tails_idx[-1]
    while idx != -1:
        seq.append(nums[idx])
        idx = predecessors[idx]

    return len(tails), seq[::-1]

# Ejemplo
nums = [10, 9, 2, 5, 3, 7, 101, 18]
length, sequence = lis_with_sequence(nums)
print(f"Longitud: {length}")
print(f"Secuencia: {sequence}")
```
