```python
import random
from collections import defaultdict

def karger_min_cut(graph):
  parent = {v: v for v in graph}

  edges = []
  for u in graph:
    for v in graph[u]:
      if u < v:
        edges.append([u,v])

  edges = [list(e) for e in edges]
  n = len(graph)

  while n > 2:
    idx = random.randint(0, len(edges)-1)

    u, v = edges[idx]

    for e in edges:
      if e[0] == v:
        e[0] == u
      if e[1] == v:
        e[1] = u

    edges = [e for e in edges if e[0] != e[1]]
    n -= 1

  return len(edges)



def global_min_cut(graph):
  import math
  n = len(graph)
  iterations = int(n*n*math.log(n)) if n > 1 else 1
  min_cut = float('inf')
  for _ in range(iterations):
    cut = karger_min_cut(graph)
    min_cut = min(min_cut, cut)
  return min_cut


if __name__ == "__main__":
  graph = {
      0: [1, 2, 3],
        1: [0, 2, 3],
        2: [0, 1, 4],
        3: [0, 1, 5],
        4: [2, 5, 6],
        5: [3, 4, 6],
        6: [4, 5],
  }

  resultado = global_min_cut(graph)
  print(f"Corte minimo global: {resultado}")
```
