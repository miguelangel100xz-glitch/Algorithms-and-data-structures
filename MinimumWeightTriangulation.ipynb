```python
import sys

def minimum_weight_triangulation(points):
  """
  Minimum Weight triangulation of convex polygon
  """

  import math

  def dist(i,j):
    x1, y1 = points[i]
    x2, y2 = points[j]
    return math.sqrt((x1-x2)** 2 + (y1-y2)**2)


  n = len(points)
  INF = float('inf')

  dp = [[0.0] * n for _ in range(n)]
  split = [[-1]*n for _ in range(n)]


  for length in range(2,n):
    for i in range(n-length):
      j = i + length
      dp[i][j] = INF
      for k in range(i+1,j):
        cost = dp[i][k] + dp[k][j] + dist(i,k) + dist(k , j) + dist(i,j)
        if cost < dp[i][j]:
          dp[i][j] = cost
          split[i][j] = k


  def reconstruct(i,j):
    if j - i < 2:
      return[]
    k = split[i][j]
    triangles = [(i,k,j)]
    triangles += reconstruct(i,k)
    triangles += reconstruct(k,j)
    return triangles

  total_cost = dp[0][n-1]
  triangles = reconstruct(0, n-1)

  return total_cost, triangles


points = [
      (0,0),
      (1,0),
      (2,1),
      (1,2),
      (0,2),
  ]


cost, triangles = minimum_weight_triangulation(points)
print(f"costo minimo: {cost: .4f}")
print(f"Triangulos (indices):")

for t in triangles:
    a,b,c = t
    print(f" {t} -> {points[a]}, {points[b]} , {points[c]}")
```
