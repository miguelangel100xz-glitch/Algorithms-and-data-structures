```python
import heapq

def prim(grafo, inicio):
  """
  Algoritmo de prim para arbol de expansion minima
  """

  visitados = set()
  aristas_arbol = []
  total_peso = 0

  heap = [(0, inicio, inicio)]

  while heap:
    peso, origen, nodo = heapq.heappop(heap)

    if nodo in visitados:
      continue

    visitados.add(nodo)
    total_peso += peso

    if nodo != origen:
      aristas_arbol.append((peso, origen, nodo))

    for (w, vecino) in grafo.get(nodo, []):
      if vecino not in visitados:
        heapq.heappush(heap,(w,nodo,vecino))

  return aristas_arbol, total_peso



grafo = {
    'A': [(2, 'B'), (3, 'C')],
    'B': [(2, 'A'), (1, 'C'), (5, 'D')],
    'C': [(3, 'A'), (1, 'B'), (2, 'D')],
    'D': [(1, 'B'), (5, 'C')],
}


aristas, peso_total = prim(grafo, 'A')
print("Aristas del árbol de expansión mínima:")
for arista in aristas:
    print(arista)
print("Peso total:", peso_total)
```
