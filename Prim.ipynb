```python
import heapq

def prim(grafo, inicio):
  """
  grafo: {nodo: [(vecino, peso), ...]}
  Devuelve (cost_total, aristas_MST)
  """

  visitado = set()
  aristas_mst = []
  costo_total = 0

  heap = [(0,inicio, None)]
  while heap:
    peso, u , origen = heapq.heappop(heap)

    if u in visitado:
      continue

    visitado.add(u)
    if origen is not None:
      aristas_mst.append((origen, u, peso))
      costo_total += peso

    for vecino, w in grafo[u]:
      if vecino not in visitado:
        heapq.heappush(heap, (w, vecino, u))

  return costo_total, aristas_mst


grafo = {
    'A': [('B',4), ('E',2)],
    'B': [('A',4), ('C',6), ('E',3), ('F',5)],
    'C': [('B',6), ('D',3), ('G',2)],
    'D': [('C',3), ('G',4)],
    'E': [('A',2), ('B',3), ('F',1)],
    'F': [('B',5), ('E',1), ('G',7)],
    'G': [('C',2), ('D',4), ('F',7)],
}


costo, aristas = prim(grafo, 'A')
print(f"Costo total: {costo}")
print(f"Aristas del MST: {aristas}")
```
