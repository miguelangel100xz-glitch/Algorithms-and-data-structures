```python
import random

def quicksort_aleatorizado(arr):
  if len(arr) <= 1:
    return arr


  pivot_idx = random.randint(0, len(arr) - 1)
  pivot = arr[pivot_idx]

  menores = [x for i, x in enumerate(arr) if x < pivot]
  iguales = [x for x in arr if x == pivot]
  mayores = [x for i, x in enumerate(arr) if x > pivot]

  return quicksort_aleatorizado(menores) + iguales + quicksort_aleatorizado(mayores)


datos = [64, 34, 25, 12, 22, 11, 90, 45]
print("Original: ", datos)
print("Ordenado: ", quicksort_aleatorizado(datos))
```
