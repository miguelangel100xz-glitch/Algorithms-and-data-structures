```python
import random
from collections import defaultdict

def karger_min_cut(grafo_original):
  """
  """

  grafo = defaultdict(list)
  for u in grafo_original:
    for v in grafo_original[u]:
      grafo[u].append(v)

  padre = {u: u for u in grafo}

  def encontrar(x):
    while padre[x] != x:
      padre[x] = padre[padre[x]]
      x = padre[x]
    return x

  def unir(x,y):
    padre[encontrar(x)] = encontrar(y)

  nodos = list(grafo.keys())
  n = len(nodos)

  while n > 2:
    u = random.choice(list(grafo.keys()))
    v = random.choice(grafo[u])

    ru, rv = encontrar(u), encontrar(v)
    if ru == rv:
      continue

    unir(ru, rv)
    n -= 1

    nuevo_grafo = defaultdict(list)
    for x in grafo:
      for y in grafo[x]:
        if encontrar(x) != encontrar(y):
          nuevo_grafo[encontrar(x)].append(encontrar(y))
    grafo = nuevo_grafo

  aristas_corte = sum(len(v) for v in grafo.values()) // 2
  return aristas_corte


def min_cut(grafo, repeticiones=None):
  n = len(grafo)
  if repeticiones is None:
    repeticiones = n * (n-1)

  mejor = float('inf')
  for _ in range(repeticiones):
    corte = karger_min_cut(grafo)
    mejor = min(mejor, corte)

  return mejor


if __name__ == "__main__":

  grafo = {
      0: [1, 2, 3],
      1: [0, 2],
      2: [0, 1, 4],
      3: [0, 4, 5],
      4: [2, 3, 5],
      5: [3, 4],
  }

  random.seed(42)
  n = len(grafo)
  reps = n * (n-1)

  print(f"Nodos     : {n}")
  print(f"Repeticiones: {reps}")

  resultado = min_cut(grafo, repeticiones=reps)
  print(f"Mejor corte: {resultado}")

  # Mostrar distribución de resultados en 200 intentos
  print("\nDistribución de 200 ejecuciones individuales:")
  conteo = defaultdict(int)
  for _ in range(200):
      conteo[karger_min_cut(grafo)] += 1
  for corte in sorted(conteo):
      barra = "█" * conteo[corte]
      print(f"  corte={corte}  {barra}  ({conteo[corte]} veces)")
```
