```python
import asyncio
import heapq

async def dijkstra_async(
    grafo: dict[int, list[tuple[int, int]]],
    origen: int,
    delay: float = 0.0

) -> tuple[dict, dict]: # Added the missing colon here

    dist = {u: float('inf') for u in grafo} # Fixed indentation and 'U' to 'u'
    dist[origen] = 0
    prev = {u: None for u in grafo}

    heap = [(0, origen)]

    while heap:
        d, u = heapq.heappop(heap)
        if d > dist[u]: # Fixed 'U' to 'u'
            continue

        vecinos = await fetch_vecinos(grafo, u, delay)

        for v, peso in vecinos:
            nueva_dist = dist[u] + peso
            if nueva_dist < dist[v]:
                dist[v] = nueva_dist
                prev[v] = u
                heapq.heappush(heap, (nueva_dist, v))

    return dist, prev # Fixed indentation


async def fetch_vecinos(
    grafo: dict, nodo: int, delay: float
) -> list[tuple[int, int]]:

    await asyncio.sleep(delay)
    return grafo.get(nodo, [])


def reconstruir_camino(prev: dict, origen: int, destino: int) -> list[int]:
  camino, nodo = [], destino
  while nodo is not None:
    camino.append(nodo)
    nodo = prev[nodo]
  camino.reverse()
  return camino if camino[0] == origen else []


async def main():
    grafo = {
        0: [(1, 3), (2, 5)],
        1: [(2, 1), (3, 7)],
        2: [(3, 2)],
        3: [(4, 4)],
        4: []
    }

    print("===Dijkstra secuencial")
    dist, prev = await dijkstra_async(grafo, origen=0, delay=0.01) # Fixed 'dealy' to 'delay'
    for nodo, d in dist.items(): # Fixed indentation
        camino = reconstruir_camino(prev, 0, nodo)
        print(f"  0 → {nodo}: dist={d}  camino={camino}")

await main()
```
