```python
def subset_sum_scheduling(tasks: list[int], W: int) -> tuple[bool, list[int]]:
    n = len(tasks)


    dp = [[False] * (W + 1) for _ in range(n+1)]
    dp[0][0] = True


    for i in range(1, n+1):
      duracion = tasks[i - 1]
      for w in range(W + 1):
        dp[i][w] = dp[i - 1][w]
        if w >= duracion and dp[i - 1][w - duracion]:
          dp[i][w] = True

    seleccionadas = []
    if dp[n][W]:
      w = W
      for i in range(n, 0, -1):
        if not dp[i - 1][w]:
          seleccionadas.append(i - 1)
          w -= tasks[i - 1]

    return dp[n][W], list(reversed(seleccionadas))

tareas = [3,5,2,7,4]
W = 9

encontrado, idx = subset_sum_scheduling(tareas, W)
if encontrado:
  print(f"Solución: {[tareas[i] for i in idx]} = {W}h")
else:
  print(f"No existe subconjunt que sume {W}h")

```
