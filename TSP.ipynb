```python
import sys
from functools import lru_cache

def held_karp(dist):
    n = len(dist)
    FULL = (1 << n) - 1

    INF = float('inf')
    dp = [[INF]*n for _ in range(1 << n)]
    parent = [[-1] * n for _ in range(1 << n)]

    dp[1][0] = 0


    for mask in range(1, 1 << n):
        for u in range(n):
            if not (mask >> u & 1): continue
            # Only proceed if dp[mask][u] has been reached
            if dp[mask][u] == INF:
                continue

            for v in range(n):
                if mask >> v & 1: continue # v must not be in current mask

                new_mask = mask | (1 << v)
                new_cost = dp[mask][u] + dist[u][v]

                if new_cost < dp[new_mask][v]:
                    dp[new_mask][v] = new_cost
                    parent[new_mask][v] = u

    best, last = INF, -1
    for u in range(1,n):
        cost = dp[FULL][u] + dist[u][0]
        if cost < best:
            best, last = cost, u

    path, mask, cur = [], FULL, last
    while cur != -1:
        path.append(cur)
        prev = parent[mask][cur]
        mask ^= (1 << cur)
        cur = prev
    path.reverse()
    path.append(0)

    return best, path


dist = [
    [0,10,15,20],
    [10,0,35,25],
    [15,35,0,30],
    [20,25,30,0]

]


cost, route = held_karp(dist)
print(cost)
print(route)
```
