```python
from collections import deque

def kahn(grafo, nodos):
  grado = {n: 0 for n in nodos}
  for u in grafo:
    for v in grafo[u]:
      grado[v] += 1

  cola = deque(n for n in nodos if grado[n]==0)
  orden = []

  while cola:
    u = cola.popleft()
    orden.append(u)
    for v in grafo.get(u, []):
      grado[v] -= 1
      if grado[v] == 0:
        cola.append(v)

  if len(orden) != len(nodos):
    raise ValueError("El grafo tiene un ciclo")

  return orden
```

```python
def topo_dfs(grafo, nodos):
  visitado = set()
  pila = []

  def dfs(u):
    visitado.add(u)
    for v in grafo.get(u, []):
      if v not in visitado:
        dfs(v)
    pila.append(u)

  for n in nodos:
    if n not in visitado:
      dfs(n)
  return pila[::-1]



grafo = {
    'A':['B','E'],
    'B':['C'],
    'C':['D','E'],
    'D':['F'],
    'E':['F'],
    'F':[],
}
nodos = ['A','B','C','D','E','F']

print(kahn(grafo, nodos))
print(topo_dfs(grafo, nodos))
```
