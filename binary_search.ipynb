```python
import math
import matplotlib.pyplot as plt


def busqueda_binaria(lista, objetivo):
  izquierda, derecha = 0, len(lista) - 1
  comparaciones = 0
  while izquierda <= derecha:
    comparaciones += 1
    medio = (izquierda + derecha) // 2
    if lista[medio] == objetivo:
      return comparaciones
    elif lista[medio] < objetivo:
      izquierda = medio + 1
    else:
      derecha = medio - 1
  return comparaciones

```

```python
tamanios = [2 **i for i in range(1,21)]
comparaciones_binaria = []
comparaciones_teorica = []

for n in tamanios:
  lista = list(range(n))
  objetivo = n - 1
  comps = busqueda_binaria(lista, objetivo)
  comparaciones_binaria.append(comps)
  comparaciones_teorica.append(math.log2(n))

  plt.figure(figsize=(10, 6))
plt.plot(tamanios, comparaciones_binaria, 'o-', color='steelblue', label='Comparaciones reales (peor caso)')
plt.plot(tamanios, comparaciones_teorica, '--', color='tomato', label='log₂(n) teórico')
plt.xscale('log', base=2)
plt.xlabel('Tamaño del arreglo (n)')
plt.ylabel('Comparaciones')
plt.title('Complejidad real de la búsqueda binaria')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('busqueda_binaria.png', dpi=150)
plt.show()
```
